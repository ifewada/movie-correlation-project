```python
# Importing the packages 
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None
```

```python
# Importing the data to be analysed
df = pd.read_csv(r'C:\Users\iwada\Downloads\Movie Data.csv')
```

```python
# Displaying the data
df
```

```python
# Data Cleaning
# Let's check for any missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# Checking for Data Types for the Columns

print(df.dtypes)
```

```python
# Drop the null values since they are negligible

df = df.dropna()
```

```python
# Check again for any missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# Change data types for some columns

df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')

df.head()
```

```python
# Split the date and country in the released column to change the year to datetime

df.released[:10]
```

```python
type(df.released[0])
```

```python
# Change the data types of the date part of the released column to datetime
df['released'] = pd.to_datetime(df['released'].str.split(' \(').str[0], errors='coerce')
```

```python
df.head()
```

```python
# Create correct Year column from released column and convert to string

df['yearcorrect'] = df['released'].astype(str).str[:4]

df.head()
```

```python
# To display the whole data

pd.set_option('display.max_rows', None)
```

```python
# Order the data by gross in descending order

df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# Check for duplicates

df.duplicated().sum()
```

```python
# Check the whole clean data

df.head()
```

```python
# Checking for Outliers

df.boxplot(column=['gross'])
```

```python
# Let's find out if there is a correlation between budget and gross
# Plot budget vs gross using seaborn
plt.title("Budget vs Gross Earnings")
sns.regplot(x="gross", y="budget", data=df, scatter_kws={"color": "red"}, line_kws={"color":"blue"})
```

```python
sns.regplot(x="score", y="gross", data=df)
```

```python
# Correlation Matrix between all numeric columns
# Filter DataFrame to include only numeric columns
numeric_df = df.select_dtypes(include='number')

# Compute correlation matrix using Pearson method
correlation_matrix = numeric_df.corr(method='pearson')
numeric_df.corr(method ='pearson')
```

```python
# Compute correlation matrix using Kendall method
numeric_df.corr(method ='kendall')
```

```python
# Compute correlation matrix using Spearman method
numeric_df.corr(method ='spearman')
```

There is high correlation between budget and gross

There is high correlation between votes and gross

```python
# Visualizing the correlation

correlation_matrix = numeric_df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title("Correlation Matrix for Numeric Features")

plt.xlabel("Movie Features")

plt.ylabel("Movie Features")

plt.show()
```

```python
# Convert the columns that are objects to category type to include in the correlation matrix

df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes


df_numerized.head()
```

```python
# Show the correlation of all the features

df_numerized.corr()
```

```python
# Show the correlation pairs

correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
# Sort the correlation pairs

sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
# Show the pairs with high correlation

high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```

# Result
Votes and budget have the highest correlation to gross earnings.
